# Analysis

```python
corr_matrix=data.corr(method='pearson').round(2)
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(corr_matrix, annot=True)
plt.show()
```

Correlation between variables:

- there is no strong correlation between variables

```python
plt.figure(figsize=(15,15))

plt.subplot(3, 2, 1)
sns.histplot(data['arrival_date']) 

plt.subplot(3, 2, 2)
sns.histplot(data['lead_time'])

plt.subplot(3, 2, 3)
sns.histplot(data['no_of_previous_cancellations'])

plt.subplot(3, 2, 4)
sns.histplot(data['no_of_previous_bookings_not_canceled'])

plt.subplot(3, 2, 5)
sns.histplot(data['avg_price_per_room'])

plt.tight_layout()
plt.show()
```

Distribution of continuous data:

- lead time: generally not a lot of time between time of booking and stay. The longer the lead time the less bookings

```python
plt.figure(figsize=(20,25))

plt.subplot(4,2,1)
sns.countplot(x = 'no_of_adults', data = data)

plt.subplot(4,2,2)
sns.countplot(x = 'no_of_children', data = data)

plt.subplot(4,2,3)
sns.countplot(x = 'no_of_weekend_nights', data = data)

plt.subplot(4,2,4)
sns.countplot(x = 'no_of_week_nights', data = data)

plt.subplot(4,2,5)
sns.countplot(x = 'type_of_meal_plan', data = data)

plt.subplot(4,2,6)
sns.countplot(x = 'required_car_parking_space', data = data)

plt.subplot(4,2,7)
sns.countplot(x = 'room_type_reserved', data = data)

plt.subplot(4,2,8)
sns.countplot(x = 'arrival_year', data = data)

plt.tight_layout()
plt.show()
```

- most reservations are for 2 adults and no children
- most bookings are for short term stay and do not require car parking
- most popular room and meal plan are type 1
- most data is from 2018

```python
plt.figure(figsize=(20,24))

plt.subplot(4,2,1)
sns.countplot(x = 'arrival_month', data = data)

plt.subplot(4,2,2)
sns.countplot(x = 'arrival_date', data = data)

plt.subplot(4,2,3)
sns.countplot(x = 'market_segment_type', data = data)

plt.subplot(4,2,4)
sns.countplot(x = 'repeated_guest', data = data)

plt.subplot(4,2,5)
sns.countplot(x = 'no_of_previous_cancellations', data = data)

plt.subplot(4,2,6)
sns.countplot(x = 'no_of_previous_bookings_not_canceled', data = data)

plt.subplot(4,2,7)
sns.countplot(x = 'no_of_special_requests', data = data)

plt.subplot(4,2,8)
sns.countplot(x = 'booking_status', data = data)


plt.tight_layout()
plt.show()
```

- the least bookings are made for January but gradually goes up until October
- most bookings are made online

```python
plt.figure(figsize = (20, 25))

plt.subplot(5,2,1)
sns.countplot(x = 'booking_status', hue = 'no_of_adults', data = data)

plt.subplot(5,2,2)
sns.countplot(x = 'booking_status', hue = 'no_of_children', data = data)

plt.subplot(5,2,3)
sns.countplot(x = 'booking_status', hue = 'no_of_weekend_nights', data = data)

plt.subplot(5,2,4)
sns.countplot(x = 'booking_status', hue = 'market_segment_type', data = data)

plt.subplot(5,2,5)
sns.countplot(x = 'booking_status', hue = 'type_of_meal_plan', data = data)

plt.subplot(5,2,6)
sns.countplot(x = 'booking_status', hue = 'required_car_parking_space', data = data)

plt.subplot(5,2,7)
sns.countplot(x = 'booking_status', hue = 'room_type_reserved', data = data)

plt.subplot(5,2,8)
sns.countplot(x = 'booking_status', hue = 'arrival_year', data = data)

plt.subplot(5,2,9)
sns.countplot(x = 'booking_status', hue = 'repeated_guest', data = data)

plt.subplot(5,2,10)
sns.countplot(x = 'booking_status', hue = 'no_of_special_requests', data = data)

plt.tight_layout()
plt.show()
```

- repeated guests are more likely not to cancel
- the more special requests the less likely the customer will cancel

```python
plt.figure(figsize = (20, 25))

plt.subplot(3,2,1)
sns.kdeplot(x='lead_time', hue='booking_status', shade=True, data=data)

plt.subplot(3,2,2)
sns.kdeplot(x='arrival_year', hue='booking_status', shade=True, data=data)

plt.subplot(3,2,3)
sns.kdeplot(x='arrival_month', hue='booking_status', shade=True, data=data)

plt.subplot(3,2,4)
sns.kdeplot(x='arrival_date', hue='booking_status', shade=True, data=data)

plt.subplot(3,2,5)
sns.kdeplot(x='avg_price_per_room', hue='booking_status',  shade=True, data=data)

plt.tight_layout()
plt.show()
```

- the shorter the lead time the less likely the booking will be cancelled
- there seems to be less cancellations between October and December
- more expensive rooms seem to get cancelled more
